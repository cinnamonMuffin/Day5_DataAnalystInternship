```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset

df = pd.read_csv("train.csv")
```

```python
#Initial checks

df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
# Printing basic statistics

df.describe(include="all")
```

```python
#Heatmap of missing values

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values in Titanic Dataset")
plt.show()

#Observation: Some data is unknown in Age, Cabin and Embarked
```

```python
#Count Plots for each attribute
```

```python
#Survival count

plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()

#The count of people who did not survive is way more than the count of those who did.
```

```python
#Passenger class Distribution chart

plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", data=df)
plt.title("Passenger Class Distribution")
plt.show()

#Observation: Most passengers were from Passenger class 3.
```

```python
#Gender Distribution

plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Gender Distribution")
plt.show()

#Observation: Among all those on board, there were almost twice as many men than women.
```

```python
#Embarkation ports

plt.figure(figsize=(6,4))
sns.countplot(x="Embarked", data=df)
plt.title("Port of Embarkation")
plt.show()

#Observation: Most passengers embarked from Port S
```

```python
#Survival rates by groups
```

```python
#gender

plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Probability")
plt.show()

#Observation: The survival rate among women is more than that of men by a large margin.
```

```python
#by Passenger class

plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Probability")
plt.show()
```

```python
#By port of embarkation

plt.figure(figsize=(6,4))
sns.barplot(x="Embarked", y="Survived", data=df)
plt.title("Survival Rate by Port of Embarkation")
plt.ylabel("Survival Probability")
plt.show()
```

```python
#Age, Fare, and their relation to survival
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Histogram of age

plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

#Observation: Most passengers on board were around the 20-40 age range
```

```python
# 2. Histogram of fare

plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True, color='salmon')
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

#Observation: Most passengers paid less than $100. Some outliers exist
```

```python
# 3. Boxplot of Age vs survival

plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2',hue=None)
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()

#Observation: Children had more chance of survival than the elderly 
```

```python
# 4. Boxplot of Fare vs survival

plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set3')
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show

#Observation: Higher fare prices are associated with higher survival
```

```python
# 5. Pairplot for numeric variables coloured by survival

sns.pairplot(df[['Age','Fare','SibSp','Parch','Survived']], hue='Survived', palette='coolwarm')
plt.suptitle("Pairplot of Numeric Features by Survival", y=1.02)
plt.show()
```
